```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import torch
import torchvision
from PIL import Image
```

```python
img_pil = Image.open('/home/trojan/Downloads/test_img.png')
img_pil = img_pil.convert('RGB')
print(img_pil.size)
```

```python
plt.hist(np.array(img_pil).ravel(), bins=50, density=True);
plt.xlabel("pixel values")
plt.ylabel("relative frequency")
plt.title("distribution of pixels")
plt.show()
```

```python
transform = torchvision.transforms.Compose([
    torchvision.transforms.ToTensor(),
    torchvision.transforms.Normalize(
        mean=[0.5],
        std=[0.5],
    ),
])
normalized_img = transform(img_pil)
```

```python
plt.hist(normalized_img.numpy().ravel(), bins=50, density=True)
plt.xlabel("pixel values")
plt.ylabel("relative frequency")
plt.title("distribution of pixels")
plt.show()
```

```python
arr = normalized_img.numpy().ravel()
print(np.amin(arr), np.amax(arr))
```
